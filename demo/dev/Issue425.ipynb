```python
from PETsARD import Config, Status

cfg = Config(filename='exp7.yaml')
seq = cfg.sequence
sts = Status(config=cfg)
test: dict = {}


while cfg.config.qsize() > 0:
    ops = cfg.config.get()
    module = cfg.module_flow.get()
    expt = cfg.expt_flow.get()

    print(f"Now is {module} with {expt}...")
    print(f"premoudle is {sts.get_pre_module(module)}")
    ops.run(ops.set_input(status=sts))


    sts.put(module, expt, ops)

    if module == 'Preprocessor':
        test[f"preproc_ops"] = ops.get_result().copy()
        test[f"preproc_sts"] = sts.get_result('Preprocessor').copy()
    elif module == 'Synthesizer':
        test[f"syn_ops_{expt}"] = ops.get_result().copy()
        test[f"syn_sts_{expt}"] = sts.get_result('Preprocessor').copy()
```
