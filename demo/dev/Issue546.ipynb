```python
import os
import pprint
import sys


path_cwd = os.getcwd()
print(path_cwd)
path_petsard = os.path.dirname(os.path.dirname(path_cwd))
print(path_petsard)
sys.path.append(path_petsard)

pp = pprint.PrettyPrinter(depth=2)
```

```python
import yaml

from PETsARD import Executor


yaml_text: str = """---
Loader:
    national2018:
        filepath: 'benchmark://nist-national2018'
        na_values:
            'N'
        column_types:
            category:
                - PUMA
                - SEX
                - MSP
                - HISP
                - RAC1P
                - NOC
                - NPF
                - HOUSING_TYPE
                - OWN_RENT
                - INDP
                - INDP_CAT
                - EDU
                - PINCP_DECILE
                - DVET
                - DREM
                - DPHY
                - DEYE
                - DEAR
                - PWGTP
                - WGTP
Splitter:
    p0.8:
        train_split_ratio: 0.8
        random_state: 123
Preprocessor:
    onehot:
        encoder:
            PUMA: encoder_onehot
            SEX: encoder_onehot
            MSP: encoder_onehot
            HISP: encoder_onehot
            RAC1P: encoder_onehot
            NOC: encoder_onehot
            NPF: encoder_onehot
            HOUSING_TYPE: encoder_onehot
            OWN_RENT: encoder_onehot
            INDP: encoder_onehot
            INDP_CAT: encoder_onehot
            EDU: encoder_onehot
            PINCP_DECILE: encoder_onehot
            DVET: encoder_onehot
            DREM: encoder_onehot
            DPHY: encoder_onehot
            DEYE: encoder_onehot
            DEAR: encoder_onehot
            PWGTP: encoder_onehot
            WGTP: encoder_onehot
        sequence:
            - 'missing'
            - 'outlier'
            - 'encoder'
            - 'scaler'
Synthesizer:
    gaussiancopula:
        method: 'sdv-single_table-gaussiancopula'
        sample_num_rows: 100
Postprocessor:
    default:
        method: 'default'
..."""

# pp.pprint(yaml_text)

cfg = yaml.safe_load(yaml_text)
pp.pprint(cfg)

yaml_file: str = f'temp_eval.yaml'
with open(yaml_file, 'w') as f:
    f.write(yaml_text)

exec = Executor(config=yaml_file)
exec.run()

os.remove(yaml_file)
```
