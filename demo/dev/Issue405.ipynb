```python
from PETsARD import Loader


load_ori: Loader = Loader(
    '20240416_exptAB_Loader[nist-ma2019]_Splitter[p80_[1-1]_[train]].csv'
)
load_syn: Loader = Loader(
    '20240416_exptAB_Loader[nist-ma2019]_Splitter[p80_[1-1]]_Preprocessor[default-sdv]_Synthesizer[sdv-copulagan]_Postprocessor[default-sdv].csv'
)
load_ori.load()
load_syn.load()
```
